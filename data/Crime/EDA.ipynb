## Crime

The Communities and Crime dataset combines socio-economicdata from the 1990 USCensus, law enforcement data from the 1990 US LEMAS survey, and crime data from the 1995FBI UCR. It was created by Michael Redmond and is provided by the UCI machine learning repos-itory [65]. The dataset contains 128 attributes such as county, population, per capita income, andnumber of immigrants. **The task consists of predicting whether the number of violent crimes perpopulation for a given community is above or below the median.** 

We consider race as the protectedattribute, which **we set to 1 if the percentage of white people divided by 5 is smaller than the percent-age of black, asian, and hispanic individuals, and to 0 otherwise**. 

We keep the following 6 features: **racePctWhite, pctWInvInc, PctFam2Par, PctKids2Par, PctYoungKids2Par, PctKidsBornNeverMar.**

```python
# Columns needed for features and cleaning

cols_for_prep = ['pctWhite', 'pctBlack', 'pctAsian', 'pctHisp', 'pctWInvInc', 
                       'PctFam2Par', 'PctKids2Par', 
                       'PctYoungKids2Par', 'PctKidsBornNeverMar', 'violentPerPop']

df = big_df[cols_for_prep].copy()
```

```python
df['violentPerPop']
```

```python
# Remove columns with unknowns

df = df[~(df['violentPerPop'] == '?')].astype(float).copy()

df.loc[:, 'ExcessCrime'] = (df['violentPerPop'] > df.median(axis=0).violentPerPop).astype(int)
```

```python
# Create the sensitive attribute (Race)

df.loc[:, 'Race'] = 0  # 0 is white, 1 is non-white
non_white_idx = df['pctWhite']/5 < (df['pctBlack'] + df['pctAsian'] + df['pctHisp'])
df.loc[non_white_idx, 'Race'] = 1

sum(df['Race']/len(df['Race']))  # Not quite balanced. Is this a problem? 
```

```python
columns_of_interest = ['pctWhite', 'pctWInvInc', 
                       'PctFam2Par', 'PctKids2Par', 
                       'PctYoungKids2Par', 'PctKidsBornNeverMar', 
                       'Race', 'ExcessCrime']
df[columns_of_interest]
```

```python
data_columns = ['pctWhite', 'pctWInvInc', 'PctFam2Par', 'PctKids2Par', 'PctYoungKids2Par', 'PctKidsBornNeverMar']

data_tensor = torch.tensor(df[data_columns].values).float()
data_mean = data_tensor.mean()
data_std = data_tensor.std()
data_tensor = (data_tensor - data_mean)/data_std
context_tensor = torch.tensor(df['Race'].values).unsqueeze(-1).float()
labels_tensor = torch.tensor(df['ExcessCrime'].values).unsqueeze(-1).float()

dset = {"data": data_tensor, "data_mean": data_mean, "data_std": data_std,
        "context": context_tensor, "labels": labels_tensor}
torch.save(dset, "data.pt")
```
